```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
print(titanic.head())
```

```python
print(titanic.info())
```

```python
print(titanic.describe(include='all'))
```

```python
print(titanic.isnull().sum())
```

```python
titanic.hist(bins=30, figsize=(15, 10), grid=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(titanic.select_dtypes(include='category').columns, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=titanic, x=column)
    plt.title(f'Count plot of {column}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Age vs Fare Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='pclass', style='survived')
plt.title('Age vs Fare by Pclass Scatter Plot')
plt.show()
```
